```python
from Main import Main
```

```python
#first parameter number of players, second parameter chest limit (required for more than one player, so the arrays have the same size)
game = Main(1000,3500) 
```

```python
import numpy as np
import pandas as pandaInstance
        

sniperLevel = {}
shotgunLevel = {}
pistolLevel = {}
assaultLevel = {}
lightArmourLevel = {}
mediumArmourLevel = {}
heavyArmourLevel = {}

index = 1

for i in game.players:
    key = "player"+ str(index)
    index += 1
    sniperLevel[key] = i.deck.sniperLevelHistory
    shotgunLevel[key] = i.deck.shotgunLevelHistory
    pistolLevel[key] = i.deck.pistolLevelHistory
    assaultLevel[key] = i.deck.assaultLevelHistory
    lightArmourLevel[key] = i.deck.lightArmourLevelHistory
    mediumArmourLevel[key] = i.deck.mediumArmourLevelHistory
    heavyArmourLevel[key] = i.deck.heavyArmourLevelHistory
    
sniperDataFrame = pandaInstance.DataFrame(sniperLevel)
shotgunDataFrame = pandaInstance.DataFrame(shotgunLevel)
pistolDataFrame = pandaInstance.DataFrame(pistolLevel)
assaultDataFrame = pandaInstance.DataFrame(assaultLevel)
lightDataFrame = pandaInstance.DataFrame(lightArmourLevel)
mediumDataFrame = pandaInstance.DataFrame(mediumArmourLevel)
heavyDataFrame = pandaInstance.DataFrame(heavyArmourLevel)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots(figsize=(15,10))  # Create a figure and an axes.



ax.set_xlim([0, 1000])
ax.plot(game.battleHistory, sniperDataFrame.mean(axis = 1), label='Sniper Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, shotgunDataFrame.mean(axis = 1), label='Shotgun Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, pistolDataFrame.mean(axis = 1), label='Pistol Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, assaultDataFrame.mean(axis = 1), label='Assault Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, lightDataFrame.mean(axis = 1), label='Light Armour Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, mediumDataFrame.mean(axis = 1), label='Medium Armour Level')  # Plot some data on the axes.
ax.plot(game.battleHistory, heavyDataFrame.mean(axis = 1), label='Heavy Armour Level')  # Plot some data on the axes.



ax.set_xlabel('Battles')  # Add an x-label to the axes.
ax.set_ylabel('Level')  # Add a y-label to the axes.
ax.set_title("Average Gear Level")  # Add a title to the axes.
ax.legend()  # Add a legend.
```
