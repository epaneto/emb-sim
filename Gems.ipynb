```python
from Gems import Gems
```

```python
#first parameter number of players, second parameter chest limit (required for more than one player, so the arrays have the same size)
game = Gems(1) 
```

```python
import numpy as np
import pandas as pandaInstance
        

gemHistory = {}

index = 1

for i in game.players:
    key = "player"+ str(index)
    index += 1
    gemHistory[key] = i.gemCountHistory
    
gemDataFrame = pandaInstance.DataFrame(gemHistory)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots(figsize=(15,10))  # Create a figure and an axes.


ax.plot(game.players[0].sessionHistory, gemDataFrame.mean(axis = 1), label='Weapon Skills ACquired')  # Plot some data on the axes.


ax.set_xlabel('Battles')  # Add an x-label to the axes.
ax.set_ylabel('Skills')  # Add a y-label to the axes.
ax.set_title("Average Skills acquired")  # Add a title to the axes.
ax.legend()  # Add a legend.
```
